```python
import numpy as np
import matplotlib.pyplot as plt

def lorenz_attractor(xyz, s= 10, r= 28, b= 8/3):
    x,y,z= xyz
    e1= s*(y- x)
    e2= r*x- y- x*z
    e3= x*y- b*z
    return np.array([e1, e2, e3])
    
steps= 15000
xyz= np.zeros((steps, 3))
xyz[0]= [0, 1, 1.05]
dt= 0.01

for i in range(1, steps):
    xyz[i]= xyz[i-1]+ lorenz_attractor(xyz[i- 1])*dt
    
plt.figure(figsize= (16, 16)).add_subplot(projection= "3d").plot(*xyz.T)
```
